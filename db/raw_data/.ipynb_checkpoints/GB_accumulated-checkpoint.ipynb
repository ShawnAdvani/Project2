```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "GBvideos.csv"

# Read Purchasing File and store into Pandas data frame
GBvideos_data = pd.read_csv(file_to_load)
```

```python
print(len(GBvideos_data))
```

```python
GBvideos_sorted_byVidID = GBvideos_data.sort_values(["video_id", "views"]) 
```

```python
GBvideos_sorted_byVidID
```

```python
print(GBvideos_count_byVidID)
```

```python
GBvideos_sorted_byVidTitle = GBvideos_data.sort_values("title") 
```

```python
GBvideos_count_byVidTitle = len(GBvideos_sorted_byVidTitle["title"].unique())
print(GBvideos_count_byVidTitle)
```

```python
# dropping ALL duplicte values 
GBvideos_unique_byVidID = GBvideos_sorted_byVidID.drop_duplicates(subset ="video_id", keep = 'last', inplace = False) 
```

```python
print(len(GBvideos_unique_byVidID))
```

```python
GBvideos_unique_byVidID.head()
```

```python
# Convert the dictionary into DataFrame   
GBvideos_uniqueID_df = pd.DataFrame(GBvideos_unique_byVidID) 
```

```python
GBvideos_uniqueID_df
GBvideos_byCatID_df = GBvideos_uniqueID_df.groupby('category_id')  
```

```python
GBvideos_byCatID_df.first()
```

```python
for category_id, group in GBvideos_byCatID_df: 
    print(category_id) 
    print(group) 
    print() 
```

```python
GBvideos_Agg_byCatID_df = GBvideos_byCatID_df.aggregate(np.sum) 
GBvideos_Agg_byCatID_df
```

```python
GBvideos_Agg_byCatID_df.to_csv('GBvideos_catID.csv') 
```
