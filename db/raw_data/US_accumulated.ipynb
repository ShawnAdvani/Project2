```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "USvideos.csv"

# Read Purchasing File and store into Pandas data frame
USvideos_data = pd.read_csv(file_to_load)
```

```python
print(len(USvideos_data))
```

```python
USvideos_sorted_byVidID = USvideos_data.sort_values(["video_id", "views"]) 
```

```python
USvideos_sorted_byVidID
```

```python
USvideos_count_byVidID = len(USvideos_sorted_byVidID["video_id"].unique())
print(USvideos_count_byVidID)
```

```python
USvideos_sorted_byVidTitle = USvideos_data.sort_values("title") 
```

```python
USvideos_count_byVidTitle = len(USvideos_sorted_byVidTitle["title"].unique())
print(USvideos_count_byVidTitle)
```

```python
# dropping ALL duplicte values 
USvideos_unique_byVidID = USvideos_sorted_byVidID.drop_duplicates(subset ="video_id", keep = 'last', inplace = False) 
```

```python
print(len(USvideos_unique_byVidID))
```

```python
USvideos_unique_byVidID.head()
```

```python
# Convert the dictionary into DataFrame   
USvideos_uniqueID_df = pd.DataFrame(USvideos_unique_byVidID) 
```

```python
USvideos_uniqueID_df
USvideos_byCatID_df = USvideos_uniqueID_df.groupby('category_id')  
```

```python
USvideos_byCatID_df.first()
```

```python
for category_id, group in USvideos_byCatID_df: 
    print(category_id) 
    print(group) 
    print() 
```

```python
USvideos_Agg_byCatID_df = USvideos_byCatID_df.aggregate(np.sum) 
USvideos_Agg_byCatID_df
```

```python
USvideos_Agg_byCatID_df.to_csv('USvideos_catID.csv') 
```
