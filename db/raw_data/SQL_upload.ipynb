```python
import pandas as pd
from sqlalchemy import create_engine
import json
```

```python
US_file = "USvideos.csv"
US_df = pd.read_csv(US_file, encoding="ISO-8859-1") 
US_df = pd.DataFrame(US_df)
US_df.head()
```

```python
GB_file = "GBvideos.csv"
GB_df = pd.read_csv(GB_file, encoding="ISO-8859-1") 
GB_df.head()
```

```python
numb = []
cat = []
with open('US_category_id.json') as json_file:
    data = json.load(json_file)
    for x in data['items']:
        numb.append(x['id'])
        cat.append(x['snippet']['title'])
    df = {
        'category_id': numb,
        'category': cat
    }
    US_cat = pd.DataFrame(df)
US_cat.head()
```

```python
US_cat['category_id'] = US_cat['category_id'].astype('int64')
US_df = US_df.merge(US_cat)
US_df = US_df[['video_id', 'trending_date', 'title', 'channel_title', 'category_id', 'category',
       'publish_time', 'tags', 'views', 'likes', 'dislikes', 'comment_count',
       'thumbnail_link', 'comments_disabled', 'ratings_disabled',
       'video_error_or_removed', 'description']]
US_df.head()
```

```python
numb = []
cat = []
with open('GB_category_id.json') as json_file:
    data = json.load(json_file)
    for x in data['items']:
        numb.append(x['id'])
        cat.append(x['snippet']['title'])
    df = {
        'category_id': numb,
        'category': cat
    }
    GB_cat = pd.DataFrame(df)
GB_cat.head()
```

```python
GB_cat['category_id'] = GB_cat['category_id'].astype('int64')
GB_df = GB_df.merge(GB_cat)
GB_df = GB_df[['video_id', 'trending_date', 'title', 'channel_title', 'category_id', 'category',
       'publish_time', 'tags', 'views', 'likes', 'dislikes', 'comment_count',
       'thumbnail_link', 'comments_disabled', 'ratings_disabled',
       'video_error_or_removed', 'description']]
GB_df.head()
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/YouTube"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
US_df.to_sql(name='USdata', con=engine, if_exists='append', index=False)
GB_df.to_sql(name='INdata', con=engine, if_exists='append', index=False)
```
